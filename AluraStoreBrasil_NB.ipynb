### Importação dos dados



```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)
```

```python
loja1.head()
```

```python
loja2.head()
```

```python
loja3.head()
```

```python
loja4.head()
```

```python
loja1["loja"] = "Loja 1"
loja2["loja"] = "Loja 2"
loja3["loja"] = "Loja 3"
loja4["loja"] = "Loja 4"

lojas_consolidadas = pd.concat([loja1, loja2, loja3, loja4], ignore_index=True)

lojas_consolidadas.head()
```

```python
lojas_consolidadas.to_csv("lojas_consolidadas.csv", index=False)
```

```python
Qtde_vendas_por_loja= lojas_consolidadas.groupby(['loja'])['Produto'].count().rename("Qtde_Loja_Geral")
Qtde_vendas_por_loja.sort_values(ascending=False)
```

```python
quant_por_loja = (
    lojas_consolidadas
    .groupby(["Categoria do Produto", "loja"])["Produto"]
    .count()
    .unstack(fill_value=0)
    .sort_values(by=["Loja 1", "Loja 2", "Loja 3", "Loja 4"], ascending=False)
)

quant_por_loja
```

```python
Qtde_vendas_por_categoria = lojas_consolidadas.groupby(['Categoria do Produto'])['Produto'].count().rename("Qtde_Categoria_Geral")
Qtde_vendas_por_categoria.sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

Qtde_vendas_por_categoria.sort_values(ascending=False).plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(4,4),
    startangle=90,
    legend=False
)
plt.title("Participação de Vendas por Categoria")
plt.ylabel("")  # remove label
plt.show()
```

```python
top_categoria = (
    lojas_consolidadas
    .groupby(['Categoria do Produto'])["Produto"]
    .count()
    .rename("Quantidade")
)

precos_categoria = lojas_consolidadas.groupby("Categoria do Produto")["Preço"].mean()

df_top_categoria = pd.concat([top_categoria, precos_categoria], axis=1).dropna()

df_top_categoria["Faturamento"] = df_top_categoria["Quantidade"] * df_top_categoria["Preço"]

df_top_categoria = df_top_categoria.sort_values("Quantidade", ascending=False)

df_top_categoria
```

# 6.Análise Geográfica (Opcional)

```python
df = pd.read_csv('lojas_consolidadas.csv')

df[['Frete', 'lat', 'lon']].corr()
```

```python
df.groupby('loja')[['Frete','lat','lon']].corr().unstack()
```

```python
plt.figure(figsize=(5,5))
for loja, dados in df.groupby('loja'):
    plt.scatter(dados['lon'], dados['lat'], label=loja, alpha=0.6)

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Distribuição Geográfica das Vendas por Loja")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def classificar_regiao(lat, lon):
    if lat > -3 and lon < -45:
        return "Norte"
    elif lat > -18 and lon > -45:
        return "Nordeste"
    elif -23 <= lat <= -18:
        return "Centro-Oeste"
    elif -30 <= lat < -23 and -50 <= lon < -40:
        return "Sudeste"
    else:
        return "Sul"

lojas_consolidadas['Regiao'] = lojas_consolidadas.apply(lambda row: classificar_regiao(row['lat'], row['lon']), axis=1)
```

```python
plt.figure(figsize=(5,5))

for regiao, dados in lojas_consolidadas.groupby('Regiao'):
    plt.scatter(dados['lon'], dados['lat'], s=50, marker='x', label=regiao)

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Mapa de Vendas por Região")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

Este relatório apresenta uma análise completa do desempenho das quatro lojas da rede Alura Store, com o objetivo de apoiar o Sr. João na decisão de vender a loja com menor performance. Os dados foram consolidados a partir das bases individuais de cada loja, conforme disponibilizados no desafio, e analisados de forma integrada.

>As etapas executadas foram:

Geração de gráficos com Matplotlib e Seaborn, conforme desenvolvido no notebook:
